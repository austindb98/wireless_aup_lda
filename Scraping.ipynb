```python
import requests, time
from bs4 import BeautifulSoup
```

```python
ncaa_schools = []
for i in range(1, 24):
    soup = BeautifulSoup(
        requests.get("https://www.ncaa.com/schools-index/" + str(i)).content
    )
    table = soup.find_all("table")[0]
    for row in table.find_all("tr")[2:]:
        cells = row.find_all("td")
        school = cells[-1]
        ncaa_schools.append(school.text)
print(len(ncaa_schools))
ncaa_schools
```

```python
from googleapiclient.discovery import build

service = build(
    "customsearch", "v1", developerKey=#DEVELOPER KEY
)
service
```

```python
import pickle
from googleapiclient.errors import HttpError

pdf_results_list = []
pdf_urls = {}

for school in ncaa_schools:
    res = None
    try:
        res = (
            service.cse()
            .list(
                q=school + ' "wireless" acceptable use policy filetype:pdf',
                cx=#CX,
            )
            .execute()
        )
    except HttpError:
        print("QPM Exceeded?")
        time.sleep(60)
        res = (
            service.cse()
            .list(
                q=school + ' "wireless" acceptable use policy filetype:pdf',
                cx=#CX,
            )
            .execute()
        )
    url = res["items"][0]["link"]
    if url not in pdf_urls:
        pdf_results_list.append(res)
        pdf_urls.add(url)
        print(school)
    else:
        print(school, "already found?")
    time.sleep(0.35)

with open("google_results_pdf.pkl", "wb") as pdf_results_file:
    pickle.dump(pdf_results_list, pdf_results_file)
```
